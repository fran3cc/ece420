```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def show_bgr(img_bgr, title=None):
    """Display a BGR image using matplotlib (RGB)."""
    plt.figure()
    if title:
        plt.title(title)
    plt.imshow(img_bgr[:, :, ::-1])  # Convert BGR â†’ RGB
    plt.axis("off")
    plt.show()
```

```python
img = cv2.imread("logo.png")
if img is None:
    raise FileNotFoundError("logo.png not found in working directory.")
show_bgr(img, "Original Image")
```

```python
annotated = img.copy()
h, w = annotated.shape[:2]

side = 20
half = side // 2
center_x, center_y = w // 2, h // 2

top_left = (center_x - half, center_y - half)
bottom_right = (center_x + half, center_y + half)
blue_bgr = (255, 0, 0)  # BGR format

cv2.rectangle(annotated, top_left, bottom_right, blue_bgr, thickness=-1)
show_bgr(annotated, "Image with Blue Center Square")
```

```python
text = "Prelab7"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1.0
thickness = 2
red_bgr = (0, 0, 255)

(text_w, text_h), baseline = cv2.getTextSize(text, font, font_scale, thickness)
x = (w - text_w) // 2
top_margin = max(10, text_h + 10)
y = top_margin + text_h

cv2.putText(
    annotated, text, (x, y),
    font, font_scale, red_bgr, thickness, lineType=cv2.LINE_AA
)

show_bgr(annotated, "Final Annotated Image")
```
